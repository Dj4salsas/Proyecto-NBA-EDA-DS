```python
MJ = per_game.loc[per_game['Player'] == 'Michael Jordan', ['PTS']]
LBJ = per_game.loc[per_game['Player'] == 'Lebron James', ['PTS']]
kobe = per_game.loc[per_game['Player'] == 'Kobe Bryant', ['PTS']]


plt.figure(figsize = (20,10))
sns.distplot(MJ, label='MJ')
sns.distplot(LBJ, label='LBJ')
sns.distplot(kobe, label='Kobe')
plt.title("Points")
plt.legend()
```
