```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_games_stats=pd.read_csv('datos/allgames_stats.csv')
all_games_stats
```

```python
print("DataFrame shape: ",all_games_stats.shape)
print("DataFrame info: ",all_games_stats.info())
print("DataFrame columns: ",all_games_stats.columns)
print("how many missing in every column: ",all_games_stats.isna().sum())
print("type of each column: ",all_games_stats.dtypes)
print("numerical column stat: ", all_games_stats.describe().T)
```

```python
jordan = all_games_stats.loc[all_games_stats['Player'] == 'Michael Jordan']
jordan
```

```python
print(f"nMin: {jordan['PTS'].min()} \nMax: {jordan['PTS'].max()} \nAvg: {round(jordan['PTS'].mean())}")
```

```python
print(f"nMin: {jordan['AST'].min()} \nMax: {jordan['AST'].max()} \nAvg: {round(jordan['AST'].mean())}")
```

```python
print(f"nMin: {jordan['TRB'].min()} \nMax: {jordan['TRB'].max()} \nAvg: {round(jordan['TRB'].mean())}")
```

```python
print(f"nMin: {jordan['FG'].min()} \nMax: {jordan['FG'].max()} \nAvg: {round(jordan['FG'].mean())}")
```

```python
a4_dims = (12, 10)
fig, ax = plt.subplots(figsize= a4_dims)

jordan['Date'] = pd.to_datetime(jordan['Date'])
sns.boxplot(jordan['Date'].dt.year, jordan['PTS'], color='red').set_title('Jordan')
```

```python
fig = plt.figure(constrained_layout=True, figsize=[11, 8])
ax = fig.add_subplot()

opponents = jordan["Opp"].unique()
wins, loses = list(), list()
for opponent in opponents:
    mapped = jordan[jordan["Opp"] == opponent]["Result"].map({"W":1, "L":0})
    wins.append(mapped.sum())
    loses.append(len(mapped) - mapped.sum())

widths = np.array(wins) - np.array(loses)
colors = ["red" if width < 0 else "green" for width in widths]

ax.hlines(y=opponents, xmin=0, xmax=widths, color=colors , alpha=0.8)
ax.hlines(y=opponents, xmin=-20, xmax=np.minimum(widths, 0)-0.5, alpha=0.2, linestyles="--")
ax.vlines(x=0, ymin=-1, ymax=len(opponents), alpha=0.3, linestyles="--")
ax.scatter(y=opponents, x=widths, color=colors)

for i in range(len(opponents)):
    ax.text(y=opponents[i], x=widths[i] + (1 if widths[i] > 0 else -1), s=widths[i], fontsize=12, color=colors[i], horizontalalignment="center", verticalalignment="center")

ax.set_title('Jordan - Games')
ax.set_yticklabels(opponents, fontsize=13)

plt.show()
```
