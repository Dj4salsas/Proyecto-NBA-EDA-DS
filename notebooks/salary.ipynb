```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary= pd.read_csv(r'C:\Users\broken\Desktop\BootCamp Data Science\Proyecto EDA\Proyecto-NBA-EDA-DS\data\salaries.csv')
salary.dtypes
```

```python
print("DataFrame shape: ",salary.shape)
print("DataFrame info: ",salary.info())
print("DataFrame columns: ",salary.columns)
print("how many missing in every column: ",salary.isna().sum())
print("type of each column: ",salary.dtypes)
print("numerical column stat: ", salary.describe().T)
```

```python
salary2=salary['Salary'].str.replace("$","")
salary2
```

```python
salary3=pd.DataFrame(salary2)
salary3=salary3.rename(columns={'Salary':'Salary2'})
salary3
```

```python
salary3['Salary2']=salary3['Salary2'].astype('int')
```

```python
salary_final=pd.concat([salary,salary3],axis=1)
salary_final
salary_final2=salary_final[['Season', 'Player', 'Salary2']]
salary_final2
```

```python
jordan = salary_final2.loc[salary_final2['Player'] == 'Michael Jordan']
```

```python
print(f"nMin: {jordan['Salary2'].min()} \nMax: {jordan['Salary2'].max()} \nSalary: {round(jordan['Salary2'].sum())}")
```
